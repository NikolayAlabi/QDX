```python
import pandas as pd
```

```python
#Using the HNSC training set for testing purposes.
df1 = pd.read_csv(r"C:\Users\nalab\Documents\Research\Bathe\HNSC\PFI\ANATDUMMY\train_raw.csv", index_col=0)
df1.head()
```

```python
df1_deciles = pd.DataFrame(index=df1.index) #create an empty dataframe for the decile scaled
decile_bins = pd.DataFrame() #create an empty dataframe for the bins of each column
for column in df1.columns: #this is a for loop that takes every column,
    a,b = pd.cut(df1[column], 10,labels = False, retbins = True) #scales it,
    df1_deciles[column] = a #adds scaled values to dataframe 
    decile_bins[column] = b.tolist() #adds bin values to dataframe
```

```python
#Using the HNSC testing set for applying the bins
df2 = pd.read_csv(r"C:\Users\nalab\Documents\Research\Bathe\HNSC\PFI\ANATDUMMY\test_raw.csv", index_col=0)
df2.head()
```

```python
#take each column and scale it based on the bin values generated from before from each column. 
df2_deciles = pd.DataFrame(index=df2.index)
for column in df2.columns:
    a = pd.cut(df2[column], bins=decile_bins[column],labels = False)
    df2_deciles[column] = a

    df2_deciles.head()
```
